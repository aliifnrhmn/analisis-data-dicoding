# Data Cleaning

Dari pemeriksaan pada setiap DataFrame tidak ditemukan adanya duplikasi data. Setiap dataset memiliki beberapa bagian kosong pada kolom PM2.5, PM10, SO2, NO2, CO, O3. TEMP, PRES, DEWP, RAIN, wd, dan WSPM. Pada bagian no, year, month, day, hour, dan station tidak ada data yang kosong

Sebelum melakukan pembersihan data, contoh data dalam dataset bisa dilihat menggunakan fungsi `head()` pada DataFrame

```python
aotizhongxin_df.head(10)
```

Masing-masing kolom data memiliki keragaman data yang bermacam-macam. Setiap kolom yang memiliki nilai kosong memiliki nilai numerik kecuali kolom `wd`. Saya menggunakan interpolasi untuk mengisi bagian-bagian kosong tersebut. Saya menggunakan fungsi `interpolate()` untuk setiap DataFrame. Metode interpolasi yang saya gunakan adalah metode linear, yaitu mengisi nilai kosong secara linier berdasarkan nilai sebelum dan sesudahnya

```python
aotizhongxin_df = aotizhongxin_df.interpolate(method='linear')
changping_df = changping_df.interpolate(method='linear')
dingling_df = dingling_df.interpolate(method='linear')
dongsi_df = dongsi_df.interpolate(method='linear')
guanyuan_df = guanyuan_df.interpolate(method='linear')
gucheng_df = gucheng_df.interpolate(method='linear')
huairou_df = huairou_df.interpolate(method='linear')
nongzhanguan_df = nongzhanguan_df.interpolate(method='linear')
shunyi_df = shunyi_df.interpolate(method='linear')
tiantan_df = tiantan_df.interpolate(method='linear')
wanliu_df = wanliu_df.interpolate(method='linear')
wanshouxigong_df = wanshouxigong_df.interpolate(method='linear')
```

# Exploratory Data Analysis (EDA)

Pertanyaan bisnis yang ingin dijawab menggunakan analisis data:
1. Stasiun manakah yang rata-rata konsentrasi suatu polutannya paling tinggi?
2. Apakah ada korelasi antara polutan PM2.5 dengan rata-rata polutan lain?
5. Pada waktu kapankah rata-rata konsentrasi masing-masing polutan mencapai puncaknya? 

Untuk mempermudah menjawab pertanyaan tersebut, saya membuat DataFrame bernama `all_df` yang merupakan gabungan dari seluruh DataFrame setiap stasiun. Setelah DataFrame `all_df` dibuat, saya menggunakan fungsi `describe()` untuk melihat gambaran statistik dari DataFrame tersebut

```python
all_df = pd.concat([aotizhongxin_df, changping_df, dingling_df, dongsi_df, guanyuan_df, gucheng_df, huairou_df, nongzhanguan_df, shunyi_df, tiantan_df, wanliu_df, wanshouxigong_df], ignore_index=True)
```

```python
all_df.describe()
```

Untuk menjawab pertanyaan pertama, saya mengelompokkan seluruh data berdasarkan stasiun dan menghitung nilai maksimal dan rata-rata dari setiap polutan menggunakan fungsi `gropupby()` dan `agg()`

```python
all_df.groupby(by='station').agg({
    "PM2.5":['max', 'mean'],
    "PM10":['max', 'mean'],
    "NO2":['max', 'mean'],
    "SO2":['max', 'mean'],
    "CO":['max', 'mean'],
    "O3":['max', 'mean']
})
```

Untuk menjawab pertanyaan kedua, saya membuat DataFrame `pollutant_all_df` yang hanya berisi data dari enam jenis polutan dalam dataset. DataFrame ini akan saya gunakan untuk melihat korelasi antara masing-masing polutan

```python
pollutant_all_df = all_df[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']]
```

```python
pollutant_all_df.corr()
```

Dari tabel korelasi di atas, dapat disimpulkan bahwa ada keterkaitan erat antara polutan PM2.5 dengan polutan PM10 dan CO, keterkaitan lemah antara polutan PM2.5 dengan SO2 dan NO2, dan keterkaitan terbalik yang lemah antara polutan PM2.5 dengan polutan 03

Untuk menjawab pertanyaan ketiga, saya membuat DataFrame `pollutant_hour_df` untuk mengelompokkan masing-masing polutan berdasarkan jam pengukurannya dan menghitung rata-rata dari masing-masing polutan menggunakan fungsi `gropupby()` dan `agg()`

```python
pollutant_hour_df = all_df.groupby(by='hour').agg({
    "PM2.5":'mean',
    "PM10":'mean',
    "NO2":'mean',
    "SO2":'mean',
    "CO":'mean',
    "O3":'mean'
}).reset_index()
pollutant_hour_df
```

# Visualization and Explanatory Analysis

## Pertanyaan 1: Stasiun manakah yang rata-rata konsentrasi suatu polutannya paling tinggi?

Untuk kepentingan visualisasi, saya membuat DataFrame untuk masing-masing polutan. Masing-masing DataFrame polutan dibuat dengan mengelompokkan data berdasarkan nama stasiun menggunakan fungsi `gropupby()` dan `agg()` dan mengurutkannya berdasarkan nilai paling tinggi polutan dengan menambahkan parameter `ascending=False` pada fungsi `sort_values()`. `reset_index()` digunakan untuk membuat DataFrame tanpa index

```python
pollutant_pm25 = all_df.groupby(by='station').agg({"PM2.5":'mean'}).sort_values(by='PM2.5', ascending=False).reset_index()
pollutant_pm25.rename(columns={
    'station':'Distrik',
    'PM2.5':'Rata-Rata Konsentrasi'
}, inplace=True)
```

```python
pollutant_pm10 = all_df.groupby(by='station').agg({"PM10":'mean'}).sort_values(by='PM10', ascending=False).reset_index()
pollutant_pm10.rename(columns={
    'station':'Distrik',
    'PM10':'Rata-Rata Konsentrasi'
}, inplace=True)
```

```python
pollutant_so2 = all_df.groupby(by='station').agg({"SO2":'mean'}).sort_values(by='SO2', ascending=False).reset_index()
pollutant_so2.rename(columns={
    'station':'Distrik',
    'SO2':'Rata-Rata Konsentrasi'
}, inplace=True)
```

```python
pollutant_no2 = all_df.groupby(by='station').agg({"NO2":'mean'}).sort_values(by='NO2', ascending=False).reset_index()
pollutant_no2.rename(columns={
    'station':'Distrik',
    'NO2':'Rata-Rata Konsentrasi'
}, inplace=True)
```

```python
pollutant_co = all_df.groupby(by='station').agg({"CO":'mean'}).sort_values(by='CO', ascending=False).reset_index()
pollutant_co.rename(columns={
    'station':'Distrik',
    'CO':'Rata-Rata Konsentrasi'
}, inplace=True)
```

```python
pollutant_o3 = all_df.groupby(by='station').agg({"O3":'mean'}).sort_values(by='O3', ascending=False).reset_index()
pollutant_o3.rename(columns={
    'station':'Distrik',
    'O3':'Rata-Rata Konsentrasi'
}, inplace=True)
```

Saya menggunakan visualisasi diagram batang untuk melihat stasiun mana yang memiliki konsentrasi polutan paling tinggi untuk masing-masing polutan. Saya memvisualisasikan lima statiun teratas dengan konsenstrasi suatu polutan paling tinggi. Dengan menentukan palet warna pada visualisasi, stasiun dengan konsentrasi polutan paling tinggi dapat disorot dengan warna berbeda

```python
fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(24,18))
colors = ["#408ABF", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x='Rata-Rata Konsentrasi', y='Distrik', data=pollutant_pm25.head(5), palette=colors, ax=ax[0][0])
ax[0][0].set_ylabel(None)
ax[0][0].set_title("Polutan PM2.5", loc="center", fontsize=15)
ax[0][0].tick_params(axis ='y', labelsize=12)

sns.barplot(x='Rata-Rata Konsentrasi', y='Distrik', data=pollutant_pm10.head(5), palette=colors, ax=ax[0][1])
ax[0][1].set_ylabel(None)
ax[0][1].set_title("Polutan PM10", loc="center", fontsize=15)
ax[0][1].tick_params(axis ='y', labelsize=12)

sns.barplot(x='Rata-Rata Konsentrasi', y='Distrik', data=pollutant_so2.head(5), palette=colors, ax=ax[1][0])
ax[1][0].set_ylabel(None)
ax[1][0].set_title("Polutan SO2", loc="center", fontsize=15)
ax[1][0].tick_params(axis ='y', labelsize=12)

sns.barplot(x='Rata-Rata Konsentrasi', y='Distrik', data=pollutant_no2.head(5), palette=colors, ax=ax[1][1])
ax[1][1].set_ylabel(None)
ax[1][1].set_title("Polutan NO2", loc="center", fontsize=15)
ax[1][1].tick_params(axis ='y', labelsize=12)

sns.barplot(x='Rata-Rata Konsentrasi', y='Distrik', data=pollutant_co.head(5), palette=colors, ax=ax[2][0])
ax[2][0].set_ylabel(None)
ax[2][0].set_title("Polutan CO", loc="center", fontsize=15)
ax[2][0].tick_params(axis ='y', labelsize=12)

sns.barplot(x='Rata-Rata Konsentrasi', y='Distrik', data=pollutant_o3.head(5), palette=colors, ax=ax[2][1])
ax[2][1].set_ylabel(None)
ax[2][1].set_title("Polutan O3", loc="center", fontsize=15)
ax[2][1].tick_params(axis ='y', labelsize=12)

plt.suptitle("Konsentrasi Polutan Paling Tinggi", fontsize=30)
```

## Pertanyaan 2: Apakah ada korelasi antara polutan PM2.5 dengan polutan lain?

Saya menggunakan visualisasi _scatter plot_ yntuk melihat korelasi antara polutan PM2.5 dengan polutan lain

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(24,12))

sns.scatterplot(x='PM2.5', y='PM2.5', data=pollutant_all_df, palette=colors, ax=ax[0][0], s=10)
ax[0][0].set_title("Polutan PM2.5", loc="center", fontsize=15)

sns.scatterplot(x='PM2.5', y='PM10', data=pollutant_all_df, palette=colors, ax=ax[0][1], s=10)
ax[0][1].set_title("Polutan PM10", loc="center", fontsize=15)

sns.scatterplot(x='PM2.5', y='SO2', data=pollutant_all_df, palette=colors, ax=ax[0][2], s=10)
ax[0][2].set_title("Polutan SO2", loc="center", fontsize=15)

sns.scatterplot(x='PM2.5', y='NO2', data=pollutant_all_df, palette=colors, ax=ax[1][0], s=10)
ax[1][0].set_title("Polutan NO2", loc="center", fontsize=15)

sns.scatterplot(x='PM2.5', y='CO', data=pollutant_all_df, palette=colors, ax=ax[1][1], s=10)
ax[1][1].set_title("Polutan CO", loc="center", fontsize=15)

sns.scatterplot(x='PM2.5', y='O3', data=pollutant_all_df, palette=colors, ax=ax[1][2], s=10)
ax[1][2].set_title("Polutan O3", loc="center", fontsize=15)

plt.suptitle("Korelasi Polutan PM2.5 Dengan Polutan Lain", fontsize=30)
```

Dengan visualisasi tersebut, dapat dilihat ada keterkaitan yang kuat antara polutan PM25 dengan polutan PM10 dan C0, serta keterkaitan lemah antara polutan PM2.5 dengan polutan S02, NO2, dan O3
